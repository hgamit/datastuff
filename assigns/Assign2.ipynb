### Dataset: The people.csv dataset contains contact information about people. The attributes of the dataset are firstname, lastname, age, company name, street name, city, county, state, zipcode, phone, email, url.

##### Q1) Load the dataset into a pandas dataframe and display the first 5 lines of the dataset along with the column headings. Note that the data does not come along with the column headings, so you should be adding that to the data frame.

```python
import pandas as pd
import numpy as np

columns = ["firstname", "lastname", "age", "companyName", "streetName", "city", "county", "state", "zipcode", "phone","email", "url"]
dataset = pd.read_csv("people.csv", names=columns, index_col=False)
dataset.head()
```

#### Q2) Drop the county column.

```python
df = dataset.drop(['county'], axis=1)
df.head()
```

#### Q3) Keep only those rows that have a minimum of 4 values(Non-NaN Values), otherwise delete them.

```python
# Columnwise NaN Value count
df.isnull().sum()
```

```python
df.shape
```

```python
# Function to print rows with NaN values
nans = lambda df: df[df.isnull().any(axis=1)]
nans(df)
```

```python
# No rows deleted 
df = df.dropna(thresh=4) #remove rows with non-'NaN' value is less than 4
df.shape
```

#### Q4) Delete rows with email missing.

```python
df = df.dropna(axis=0, subset=['email']) #remove row if there is any 'NaN' value in column 'email'
df.shape
```

#### Q5) Impute the missing values in age with the mean of the column.

```python
#From the scikit.imputelibrary we first import the SimpleImputerclass
from sklearn.preprocessing import Imputer
#Next we define an object of the SimpleImputerclass by looking at the docstring (use Shift+Tab)
imputer = Imputer(missing_values=np.nan, strategy='mean')
imputer.fit(df[["age"]])
df["age"]= imputer.transform(df[["age"]])
df.isnull().sum()
```

#### Q6) Perform one hot encoding for the state column.

```python
#Replace nan from state column
df['state'].replace(np.nan, 'CA', regex=True, inplace=True)
df.isnull().sum()
```

```python
onehot = pd.get_dummies(df['state'])
df = onehot.join(df)
df.shape
```
