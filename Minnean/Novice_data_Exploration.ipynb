```python
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tls
```

```python
nov_data = pd.read_csv('./Data_Novice/novice_data.csv')

for col_name in list(nov_data):
    col_name_new = col_name.strip()
    col_name_new = re.sub(' +', ' ', col_name_new)
    #col_name_new = re.sub('.', '', col_name_new)
    col_name_new = col_name_new.replace(" ", "_")
    col_name_new = col_name_new.replace("&", "_")
    col_name_new = col_name_new.replace("__", "_")
    nov_data.rename(columns={col_name: col_name_new}, inplace=True)
    
nov_data['Date'] = pd.to_datetime(nov_data['Year'].astype(str)+' '+nov_data['Month'].astype(str))
nov_data.head()
```
